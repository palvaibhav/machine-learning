```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([1,2,4,5,6,10,12,15])
plt.plot(y) #if we don't provide the x-axis then the indices of the y values are made the x-axis
#plt.show() 

#plt.plot plots the coordinates and joins them with line
y2 = np.array([3,4,6,2,6,10])
plt.plot(y2)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,50,80,90])

plt.plot(x,y)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.show()

plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.scatter(x,y, color='red', marker='^') # scatter only plots the coordinates
plt.title("Stock Data") # gives title to the plot
plt.xlabel("Days") # gives label to x-axis
plt.ylabel("Price") # gives label to y-axis
plt.show()
```

```python
x = np.random.random((1000,2)) # generates a random matrix of 1000x2

# plot 0th col of ix row as x cord. and 1th col of ix row as y cord.
for ix in range(x.shape[0]):
    plt.scatter(x[ix,0], x[ix,1], color='green',marker='+') 
plt.show()
```
