```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
numbers = np.random.randint(0,10,5) # gives a np array of 5 random integers in range [0,9]
indices = np.arange(0,5) # arange gives a list of numbers from 0 to n-1

numbers2 = np.random.randint(0,10,5) 
indices2 = np.arange(0,5)

print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers,width=0.25,color='orange',label='Gold') # plots a bar chart
plt.bar(indices2+0.25,numbers2,width=0.25,color='red',label='Silver')
plt.legend()
#plt.bar?
plt.show()
```

```python
labels = ['Algebra','Calculus','Coding','Algorithms']
values = [90,80,54,40]

plt.pie(values,labels=labels,radius=2,explode=(0,1,0,0),shadow=True)
'''
explode: highlights any section with 1 in the corresponding tuple value
shadow: gives shadow to the chart
radius: gives radius to the chart

'''
plt.show()
```
