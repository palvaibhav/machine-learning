```python
import matplotlib.pyplot as plt
import numpy as np
```

# Univariate Case

```python
#Generate a normal distribution (1-D)

u = 5
sigma = 3
#a normal distribution is called standard normal distribution if u = 0 & sigma = 1

vals = np.random.randn(1000) 
plt.hist(vals,15) # histogram shows number of values falling in a interval
plt.title('Standard Normal Distribution')
plt.show()

vals = u + sigma * vals
plt.hist(vals,15)
plt.title('Normal Distribution')
plt.show()
'''
randn gives a standard normal distribution by default
added u to shift the curve and have mean 5
multiplied sigma for increasing the spread and have SD 2

'''
#print(vals.shape)
#print(vals)

```

```python
x = vals
y = np.zeros(vals.shape)

plt.scatter(x,y)
plt.show()
```

```python
vals = np.round(vals) #Rounding off each value to an integer
z = np.unique(vals, return_counts=True)
'''
Returns a tuple of two np arrays 
first member of tuple is an np array of unique values 
second member of tuple is an np array of frequency of coressponding unique values
'''
print(z)
```

# Multivariate Case

```python
#Bivariate
mean = np.array([0.0,0.0])    # Mean along each axis
cov = np.array([[1,0],[0,1]]) # Covariance matrix

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.1,-0.8],[-0.8,0.9]])

#Generate the distribution 
dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
'''
Give mean, covariance and number of points needed
'''
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
