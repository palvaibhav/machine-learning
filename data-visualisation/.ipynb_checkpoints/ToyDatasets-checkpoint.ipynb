```python
from sklearn.datasets import load_digits,load_boston,make_classification,make_regression,make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
mnist = load_digits()
# load_digits is basically a dataset in sklearn of images of handwritten digits
X = mnist.data # data is the X values
Y = mnist.target # target is their labels
```

```python
plt.imshow(X[3].reshape(8,8),cmap='gray')
print(Y[3])
plt.show()
```

```python
boston = load_boston()
'''
This is a data set of prices of houses in boston according to 13 features
'''
```

```python
XB = boston.data
YB = boston.target

print(XB.shape)
print(YB.shape)
```

```python
X,Y = make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,n_classes=2)
'''
n_samples : Number of data points needed
n_features : number of features they going to have
n_informative : Number of features that are informative
n_redundant : Number of features that are redundant
n_classes : The number of different classes the data points gonna have
'''
plt.figure(0,figsize=(15,15))
plt.subplot(221)
plt.title("Classification in 2 classes")
plt.scatter(X[:,0],X[:,1],c=Y,edgecolor='k')

X1,Y1 = make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,n_classes=3,n_clusters_per_class=1)
'''
n_clusters_per_class : if n_classes > 2 then we need to specify no. of clusters per class
'''
plt.subplot(222)
plt.title("Classification in 3 classes")
plt.scatter(X1[:,0],X1[:,1],c=Y1,edgecolor='k')

X3,Y3 = make_blobs(n_features=2,centers=3)
'''
centers : defines how many groups we want
'''
plt.subplot(223)
plt.title("Blobs")
plt.scatter(X3[:,0],X3[:,1],c=Y3,edgecolor='k')

X4,Y4 = make_regression(n_features=1,n_samples=200,bias=4,noise=15)
'''
bias : the intercept on Y-axis when x = 0
noise : how much the data points much deviate from the line
'''
plt.subplot(224)
plt.title("Regression")
plt.scatter(X4,Y4,color='red',edgecolor='k')

plt.show()
```
